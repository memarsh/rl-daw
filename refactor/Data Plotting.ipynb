```python
%matplotlib inline
import pandas
import pytry
import seaborn
import pylab
```

```python
dataa = pytry.read('data')
datab = pytry.read('old_data_may23')
data = dataa + datab
```

```python
data2 = []
for d in data:
    stay = d['stay_prob']
    for rewarded in [True, False]:
        for rare in [True, False]:
            d2 = dict(d)
            d2['rewarded']='+' if rewarded else '-'#rewarded
            d2['rare']='R' if rare else 'C'#rare
            d2['stay']=stay[int(rewarded),int(rare)]
            d2['type']='%s%s' % ('R' if rare else 'C', '+' if rewarded else '-')
            data2.append(d2)
            d2.pop('history')
            d2.pop('rewards')
            if 'learn_intervals' not in d2:
                if 'n_intervals' == 20000:
                    d2['learn_intervals']=20000
                else:
                    d2['learn_intervals'] = 1234
                    
df = pandas.DataFrame(data2)
df = df.drop_duplicates(subset=['seed', 'stay'])
```

```python
seaborn.set_style("whitegrid")
pylab.figure(figsize=(4,3))

colors = seaborn.color_palette('deep', 2)
palette = {
    'C+': colors[0],
    'R+': colors[1],
    'C-': colors[0],
    'R-': colors[1],
}
```
